```python
import pandas as pd

df_team = pd.read_csv('teamstats.csv')
df_shot = pd.read_csv('shots.csv')

# merge both dataframes based on gameID and drop unnecessary columns
merged_df = df_team[['gameID', 'teamID', 'shots']].merge(df_shot[['gameID', 'shotType']], on='gameID').drop(columns=['assisterID', 'minute', 'situation', 'lastAction', 'xGoal', 'positionX', 'positionY'])

# calculate the average number of shots per game for each team
avg_shots_per_game = merged_df.groupby('teamID')['shots'].mean()

# save the output in a variable
OUTPUT = avg_shots_per_game
```

```python
import pandas as pd

# Read the CSV files into dataframes
teams_df = pd.read_csv('teams.CSV')
leagues_df = pd.read_csv('leagues.CSV')

# Filter out unnecessary columns from teams and leagues dataframes
teams_df = teams_df[['teamID', 'name']]
leagues_df = leagues_df[['leagueID', 'name', 'understatNotation']]

# Merge the two dataframes on the 'teamID' column
merged_df = pd.merge(teams_df, leagues_df, left_on='teamID', right_on='leagueID')

# Extract the unique list of leagues and store it in OUTPUT variable
OUTPUT = merged_df['name'].unique().tolist()
```

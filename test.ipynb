```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('employees.csv')

# Check if 'salary' column exists in dataframe
if 'salary' in df.columns:

    # Process the 'salary' column
    df['salary'] = df['salary'].replace({'k': ''}, regex=True).astype(float) * 1000

    # Write the result to a new CSV file
    df.to_csv('new_employees.csv', index=False)

else:
    print("'salary' column not found in the CSV file.")
```

```python
from fpdf import FPDF

# Natural language input
input_text = "Generate a report titled 'Project Update' with the following content: 'The project is on track and within budget.'"

# Extract title and content (basic example, can be improved)
title = input_text.split("'")[1]
content = input_text.split("'")[3]

# PDF generation
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)

# Add title
pdf.cell(200, 10, txt=title, ln=True, align='C')

# Add content
pdf.multi_cell(0, 10, content)

# Save PDF
pdf.output("Generated_Report.pdf")
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.units import inch

# Create a PDF document
doc = SimpleDocTemplate("invoice.pdf", pagesize=letter)

# Create a list of flowables (elements to be added to the PDF)
elements = []

# Add a title to the invoice
title_style = getSampleStyleSheet()["Title"]
title = Paragraph("Invoice", title_style)
elements.append(title)

# Create a table for the invoice details
data = [
    ["Description", "Quantity", "Unit Price", "Total"],
    ["Product A", "1", "$1,200", "$1,200"],
]

# Define the table style
table_style = TableStyle([
    ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
    ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
    ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
    ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
    ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
    ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
    ('GRID', (0, 0), (-1, -1), 1, colors.black)
])

# Create the table
table = Table(data, colWidths=[2*inch, 1*inch, 1.5*inch, 1.5*inch])
table.setStyle(table_style)
elements.append(table)

# Calculate the total amount
total_amount = 1200

# Add the total amount to the invoice
total_style = getSampleStyleSheet()["Normal"]
total = Paragraph(f"<b>Total: ${total_amount}</b>", total_style)
elements.append(total)

# Build the PDF document
doc.build(elements)

print("Invoice PDF generated successfully.")
```
